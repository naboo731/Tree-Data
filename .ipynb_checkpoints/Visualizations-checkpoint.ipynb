```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Street_Tree_List.csv')
```

```python
isTree = data[data.PlantType == "Tree"]
totalTrees=(isTree.PlantType.count())
uniqueZipcodes = data['Zip Codes'].unique()
print(uniqueZipcodes)
treesByZipcode = data.groupby('Zip Codes')['TreeID']
treesByZipcode.count().plot.hist()
plt.title('Trees by Zip Code')
plt.xlabel('Zip Code')
plt.ylabel('Number of Trees')
```

```python
private = data[data.qCaretaker == "Private"]
DPW = data[data.qCaretaker == "DPW"]
caretaker = data.groupby('qCaretaker')['TreeID']
caretaker.count().plot.bar()
plt.title('Trees by Caretaker')
plt.xlabel('Caretaker')
plt.ylabel('Number of Trees')
```
